```python
import torch
import torch.nn as nn
from torch import optim
import torch.nn.functional as F
```

```python
from zipfile import ZipFile
fn = 'data - Copy.zip'

with ZipFile(fn,'r') as zip:
  zip.extractall()
```

```python
from models import U_Net
u_net = U_Net().cuda()
```

```python
from DataSetClass import RetinaDataset 
tr = torch.utils.data.DataLoader(RetinaDataset('data/training/','data/manual/'), batch_size=8, shuffle=True)
```

```python
import torch.optim as optim

optimizer = optim.Adam(u_net.parameters())
# criterion = F.BCELoss()
# criterion = nn.CrossEntropyLoss()
criterion = nn.functional.mse_loss
```

```python
loss_history = []
for epoch in range(10):
    train_loss = 0.0
    for i,data in enumerate(tr):
        x, y = data
        x, y = x.cuda(), y.cuda()
        optimizer.zero_grad()
        y_pred = u_net(x)
        loss = criterion(y_pred, y)
        x = loss.item()
        train_loss += x
        if (i%10==0):
            print('(%d, %5d) item-loss: %.9f'%(epoch + 1, i + 1, x))
        loss.backward()
        optimizer.step()   
    print('[%d] epoch-loss: %.9f'%(epoch + 1, train_loss))
    loss_history.append(train_loss)
print('Finished Training')
```

```python
import matplotlib.pyplot as plt
plt.plot(loss_history)
```

```python
##saving the model
PATH = 'retina_unet_256_16_msef3.pth'
torch.save(u_net.state_dict(), PATH)
##loading the model
model = U_Net()
model.load_state_dict(torch.load(PATH))
model.eval()
```
